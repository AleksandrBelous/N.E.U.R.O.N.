```python
import os
import re
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
import numpy as np

# Путь к папке с объединенными файлами
combined_path = "./combined/"

# Сбор данных по важности признаков и размеру выборки
feature_importances_data = []

for combined_file in os.listdir(combined_path):
    if combined_file.endswith(".csv"):
        # Чтение данных
        df = pd.read_csv(os.path.join(combined_path, combined_file))

        # Признаки (X) и целевая переменная (y)
        X = df.drop(columns=["Label"])
        y = df["Label"]

        # Обучение модели случайного леса
        model = RandomForestClassifier(random_state=0)
        model.fit(X, y)

        # Важности признаков
        importances = model.feature_importances_
        feature_names = X.columns

        # Извлечение размера выборки пакетов из названия файла
        packet_size = int(re.search(r"(\d+)_packets", combined_file).group(1))

        # Сохранение данных
        feature_importances_data.append((packet_size, feature_names, importances))

# Сортировка по длине выборки пакетов
feature_importances_data.sort(key=lambda x: x[0])

# Подготовка данных для визуализации
packet_sizes = [data[0] for data in feature_importances_data]
# Предполагается, что все файлы имеют одинаковые признаки
all_feature_names = feature_importances_data[0][1]
num_features = len(all_feature_names)

# Создание матрицы для важностей признаков
importance_matrix = np.zeros((len(packet_sizes), num_features))

# Заполнение матрицы важностей
for i, (_, feature_names, importances) in enumerate(feature_importances_data):
    importance_matrix[i, :] = importances

# Настройка цветов для разных пакетных выборок
colors = plt.cm.viridis(np.linspace(0, 1, len(packet_sizes)))

# Визуализация важности признаков на одном графике
plt.figure(figsize=(14, 8))  # Увеличиваем размер графика
bar_width = 0.1  # Ширина столбцов
x = np.arange(num_features)  # Позиции для столбцов

# Построение группированных столбцов
for i in range(len(packet_sizes)):
    plt.bar(
        x + i * bar_width,
        importance_matrix[i, :],
        width=bar_width,
        label=f"{packet_sizes[i]} packets",
        color=colors[i],
    )

plt.xlabel("Features", fontsize=14)
plt.ylabel("Feature Importance", fontsize=14)
plt.title("Feature Importances from RandomForest Across Packet Intervals", fontsize=16)
plt.xticks(
    x + bar_width * (len(packet_sizes) - 1) / 2,
    all_feature_names,
    rotation=45,
    ha="right",
    fontsize=12,
)
plt.legend(title="Packet Interval Size", fontsize=12)
plt.tight_layout()  # Автоматическая настройка для уменьшения наложения
plt.grid(
    axis="y", linestyle="--", alpha=0.7
)  # Добавление сетки по оси Y для удобства восприятия
plt.show()
```
