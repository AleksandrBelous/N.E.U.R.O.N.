```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_average_packet_size(file1, file2):
    # Загрузка данных из файлов CSV
    df1 = pd.read_csv(file1)
    df2 = pd.read_csv(file2)

    # Добавление идентификаторов строк для сохранения оригинальных индексов
    df1["Index"] = range(len(df1))
    df2["Index"] = range(len(df2))

    # Объединение данных из обоих файлов
    df = pd.concat([df1, df2], ignore_index=True)

    # Фильтрация данных по метке
    df_label_1 = df[df["Label"] == 1]
    df_label_0 = df[df["Label"] == 0]

    # Построение графика
    plt.figure(figsize=(10, 6))
    plt.scatter(
        df_label_1["Average Packet Size"],
        df_label_1["Index"],
        color="red",
        label="Label 1",
    )
    plt.scatter(
        df_label_0["Average Packet Size"],
        df_label_0["Index"],
        color="green",
        label="Label 0",
    )

    # Настройка графика
    plt.title("Average Packet Size by Label")
    plt.xlabel("Average Packet Size")
    plt.ylabel("Index")
    plt.legend()
    plt.grid(True)

    # Показать график
    plt.show()
```

```python
# Пример использования функции
plot_average_packet_size(
    "csvs/sql-inj/attack/sql-inj-tcp-ip-vi-all-16-40-16-42-349_10_1_.csv",
    "csvs/sql-inj/normal/sql-inj-tcp-ip-vi-all-16-40-16-42-775_10_0_.csv",
)
```

```python
def plot_all_columns(file1, file2):
    # Загрузка данных из файлов CSV
    df1 = pd.read_csv(file1)
    df2 = pd.read_csv(file2)

    # Добавление идентификаторов строк для сохранения оригинальных индексов
    df1["Index"] = range(len(df1))
    df2["Index"] = range(len(df2))

    # Объединение данных из обоих файлов
    df = pd.concat([df1, df2], ignore_index=True)

    # Фильтрация данных по метке
    df_label_1 = df[df["Label"] == 1]
    df_label_0 = df[df["Label"] == 0]

    # Перебор всех столбцов, кроме 'Label' и 'Index'
    for column in df.columns:
        if column not in ["Label", "Index"]:
            plt.figure(figsize=(10, 6))
            plt.scatter(
                df_label_1[column], df_label_1["Index"], color="red", label="Label 1"
            )
            plt.scatter(
                df_label_0[column], df_label_0["Index"], color="green", label="Label 0"
            )

            # Настройка графика
            plt.title(f"{column} by Label")
            plt.xlabel(column)
            plt.ylabel("Index")
            plt.legend()
            plt.grid(True)

            # Показать график
            plt.show()
```

```python
# Пример использования функции
plot_all_columns(
    "csvs/sql-inj/attack/sql-inj-tcp-ip-vi-all-16-40-16-42-349_10_1_.csv",
    "csvs/sql-inj/normal/sql-inj-tcp-ip-vi-all-16-40-16-42-775_10_0_.csv",
)
```

```python
def plot_all_columns_horizontally(file1, file2):
    # Загрузка данных из файлов CSV
    df1 = pd.read_csv(file1)
    df2 = pd.read_csv(file2)

    # Добавление идентификаторов строк для сохранения оригинальных индексов
    df1["Index"] = range(len(df1))
    df2["Index"] = range(len(df2))

    # Объединение данных из обоих файлов
    df = pd.concat([df1, df2], ignore_index=True)

    # Фильтрация данных по метке
    df_label_1 = df[df["Label"] == 1]
    df_label_0 = df[df["Label"] == 0]

    # Получаем список всех столбцов, кроме 'Label' и 'Index'
    columns = [col for col in df.columns if col not in ["Label", "Index"]]

    # Создаем подграфики в одну строку
    fig, axes = plt.subplots(1, len(columns), figsize=(5 * len(columns), 6))

    # Если у нас только один график, `axes` не будет массивом, поэтому приведем его к массиву
    if len(columns) == 1:
        axes = [axes]

    # Перебор всех столбцов и создание графиков
    for ax, column in zip(axes, columns):
        ax.scatter(
            df_label_1[column], df_label_1["Index"], color="red", label="Label 1"
        )
        ax.scatter(
            df_label_0[column], df_label_0["Index"], color="green", label="Label 0"
        )
        ax.set_title(f"{column} by Label")
        ax.set_xlabel(column)
        ax.set_ylabel("Index")
        ax.legend()
        ax.grid(True)

    # Устанавливаем плотное расположение графиков
    plt.tight_layout()

    # Показать все графики
    plt.show()
```

```python
# Пример использования функции
plot_all_columns_horizontally(
    "csvs/sql-inj/attack/sql-inj-tcp-ip-vi-all-16-40-16-42-349_10_1_.csv",
    "csvs/sql-inj/normal/sql-inj-tcp-ip-vi-all-16-40-16-42-775_10_0_.csv",
)
```
